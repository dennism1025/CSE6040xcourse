```python
import pandas as pd
data = pd.read_csv("temperatures-us-cities.csv")
data.head(10)
```

```python
atl=data.loc[data["City"] == "Atlanta"]
atl.head()
```

```python
atl.dropna (inplace=True)
atl.head (10)
```

```python
atl['year'] = ""
atl['year'] = atl.dt.str[:4]
atl.head()
```

```python
atl_means = pd.dataframe['year','AverageTemperatureC']
atl_means ['year','AverageTemperatureC']=atl.groupby(['year'])['AverageTemperatureC'].mean()
```
